# Data Preprocessing and Visualization of LLMs Data (2018-2024) Dataset

This notebook provides an in-depth preprocessing and visualization of a LLMs Data (2018-2024) dataset, which was downloaded from Kaggle ([link](https://www.kaggle.com/datasets/jainaru/llms-data-2018-2024)). The analysis utilizes several Python libraries, including `numpy`, `pandas`, `matplotlib`, and `seaborn`, to ensure comprehensive data manipulation and insightful visual representations.

## Dataset Description

This dataset contains information about every major language model and chatbot released since 2018, including details about the company that developed the model, its architecture, and various other attributes.

- **Model:** The name of the language model.
- **Company:** The company that developed the model.
- **Arch:** The architecture of the model (e.g., Transformer, RNN). "TBA" stands for To Be Announced.
- **Parameters:** The number of parameters (weights) in the model, which is a measure of its complexity. This value is provided in billions.
- **Tokens:** The number of tokens (sub-word units) the model can process or was trained on. Some values are marked as "TBA" (To Be Announced), and this value is provided in billions.
- **Ratio:** Likely the ratio of parameters to tokens, or some other relevant ratio. In this table, it is specified only for some models (e.g., Olympus as 20:01).
- **ALScore:** ALScore is a quick and dirty rating of the model's power. The formula used is the square root of (Parameters x Tokens).
- **Training dataset:** The dataset used to train the model.
- **Release Date:** The expected or actual release date of the model.
- **Notes:** Additional notes about the model, such as training details or related information.
- **Playground:** A URL link to a website where you can interact with the model or find more information about it.


## Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the dataset

```python
df = pd.read_csv('Large language models dataset.csv',encoding='unicode_escape')
```

## Data Preprocessing

```python
# Checking for missing values
df.isnull().sum()
```

```python
# Filling missing values with a placeholder
df.fillna('Unknown', inplace=True)
```

```python
# Renaming a column
df.rename(columns={'Comapany': 'Company'}, inplace=True)
df.rename(columns={'Tokens ': 'Tokens'}, inplace=True)
```

```python
# Converting numerical columns to appropriate types
df['Parameters'] = pd.to_numeric(df['Parameters'], errors='coerce')
df['Tokens'] = pd.to_numeric(df['Tokens'], errors='coerce')
df['ALScore'] = pd.to_numeric(df['ALScore'], errors='coerce')
```

```python
# Replacing 'TBA' with np.nan
df.replace('TBA', np.nan, inplace=True)
```

```python
# Dropping rows with any missing values
df.dropna(inplace=True)
```

```python
# Dropping duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Standardizing text columns to lowercase
df['Model'] = df['Model'].str.lower()
df['Company'] = df['Company'].str.lower()
```

```python
# Removing leading and trailing spaces from text columns
df['Model'] = df['Model'].str.strip()
df['Company'] = df['Company'].str.strip()
```

```python
# Categorizing ALScore into bins
bins = [0, 10, 20, 30, 40]
labels = ['Low', 'Medium', 'High', 'Very High']
df['AL_Score_Category'] = pd.cut(df['ALScore'], bins=bins, labels=labels)
```

```python
# Checking the datatypes of each column
df.dtypes
```

```python
# Getting summary statistics for numerical columns
df.describe()
```

```python
# Counting the occurrences of each model
df['Model'].value_counts()
```

```python
# Groupping by company and get mean AL Score
df.groupby('Company')['ALScore'].mean()
```

```python
# Sorting dataset by AL Score
df.sort_values(by='ALScore', ascending=False, inplace=True)
```

```python
# Normalizing numerical columns
df['Parameters'] = (df['Parameters'] - df['Parameters'].mean()) / df['Parameters'].std()
df['Tokens'] = (df['Tokens'] - df['Tokens'].mean()) / df['Tokens'].std()
```

```python
# Removing rows where 'Parameters' is NaN
df = df[~df['Parameters'].isna()]
```

## Data visualization

```python
# Plotting a correlation matrix
correlation_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Plotting a diagonal correlation matrix
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(8, 5))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Plotting Scatter plot of AL Score vs Parameters
plt.scatter(df['Parameters'], df['ALScore'])
plt.title('Scatter plot for AL Score vs Parameters')
plt.xlabel('Parameters')
plt.ylabel('ALScore')
plt.show()
```

```python
# Plotting Histogram of AL Scores
df['ALScore'].plot(kind='hist', bins=10)
plt.title('Histogram of Distribution of AL Scores')
plt.xlabel('AL Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting Line plot of Tokens over Models
top_10_df = df.nlargest(6, 'Tokens')
plt.figure(figsize=(8, 5))
top_10_df.set_index('Model')['Tokens'].plot(kind='line', marker='o')
plt.title('Line plot of Tokens per top 6 Models')
plt.xlabel('Model')
plt.ylabel('Tokens')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plotting Bar plot of AL Score categories
df['AL_Score_Category'].value_counts().plot(kind='bar')
plt.title('Bar plot of AL Score Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
# Box plot for Parameters and Tokens
plt.figure(figsize=(12, 6))
plt.boxplot([top_10_df['Parameters'], top_10_df['Tokens']], labels=['Parameters', 'Tokens'])
plt.title('Box Plot of Parameters and Tokens')
plt.ylabel('Values')
plt.show()
```

```python
# Plotting Stacked Bar Chart for Parameters and Tokens
top_10_df = df.nlargest(10, 'Tokens')
width = 0.4
models = top_10_df['Model']
parameters = top_10_df['Parameters']
tokens = top_10_df['Tokens']

fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(models, parameters, width, label='Parameters')
ax.bar(models, tokens, width, bottom=parameters, label='Tokens')
ax.set_xlabel('Model')
ax.set_ylabel('Values')
ax.set_title('Stacked Bar Chart for Parameters and Tokens per top 10 Models')
ax.legend()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plotting Area Chart for Tokens and Parameters per top 10 Models
top_10_df = df.nlargest(10, 'Tokens')
plt.figure(figsize=(10, 6))
plt.fill_between(top_10_df['Model'], top_10_df['Parameters'], alpha=0.5, label='Parameters')
plt.fill_between(top_10_df['Model'], top_10_df['Tokens'], alpha=0.5, label='Tokens')
plt.xlabel('Model')
plt.ylabel('Values')
plt.title('Area Chart of Parameters and Tokens per top 10 Models')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plotting Step Chart for Tokens
top_10_df = df.nlargest(10, 'Tokens')
plt.figure(figsize=(10, 6))
plt.step(top_10_df['Model'], top_10_df['Tokens'], where='mid')
plt.xlabel('Model')
plt.ylabel('Tokens ')
plt.title('Step Chart of Tokens per top 10 Models')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plotting Pie Chart for Model Distribution
top_10_df = df.nlargest(10, 'Tokens')
plt.figure(figsize=(10, 8))
plt.pie(top_10_df['Tokens'], labels=top_10_df['Model'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(np.arange(10)))
plt.title('Distribution of Tokens among Top 10 Models')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Plotting Density Plot of AL Scores
plt.figure(figsize=(10, 6))
df['ALScore'].plot(kind='kde')
plt.title('Density Plot of AL Scores')
plt.xlabel('AL Score')
plt.ylabel('Density')
plt.show()
```
